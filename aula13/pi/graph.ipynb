```python
# Pi Parallel For
# MIN_BLK: 1024
# for 2147483648 steps pi = 3.14159265358996 in 4.22081642784178 secs
# MIN_BLK: 512
# for 2147483648 steps pi = 3.14159265359045 in 6.9088870678097 secs
# MIN_BLK: 256
# for 2147483648 steps pi = 3.14159265359101 in 13.1699356138706 secs

# Pi Task
# MIN_BLK (task): 1024
# for 2147483648 steps pi = 3.14159265358974 in 3.2536821430549 secs
# MIN_BLK (task): 512
# for 2147483648 steps pi = 3.14159265358955 in 5.37541925907135 secs
# MIN_BLK (task): 256
# for 2147483648 steps pi = 3.14159265358981 in 11.9432846270502 secs

parallel = [4.22081642784178, 6.9088870678097, 13.1699356138706]
task = [3.2536821430549, 5.37541925907135, 11.9432846270502]
x = [1024, 512, 256]

plt.plot(x, parallel, label='Parallel For')
plt.plot(x, task, label='Task')
plt.xlabel('Block Size')
plt.ylabel('Time (s)')
plt.title('Pi Calculation Time')
plt.legend()
plt.show()
```
