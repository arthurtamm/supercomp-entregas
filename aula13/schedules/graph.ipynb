```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
labels = ['Default', 'Static', 'Dynamic', 'Guided', 'Auto']
mean = [0.000184546, 7.31967e-06, 1.72727e-05, 1.09413e-05, 7.03767e-06]
colors = ['b', 'g', 'r', 'c', 'm']
plt.bar(labels, mean, color=colors)
plt.ylabel('Tempo de execução (s)')
plt.title('Tempo de execução médio')
plt.show()
```

```python
# Variações entre execuções:
# Default:  0.000536189 segundos
# Static:   1.518e-06 segundos
# Dynamic:  6.075e-06 segundos
# Guided:   1.783e-06 segundos
# Auto:     3.78001e-07 segundos

var = [0.000536189, 1.518e-06, 6.075e-06, 1.783e-06, 3.78001e-07]
plt.bar(labels, var, color=colors)
plt.ylabel('Variação de tempo de execução (s)')
plt.title('Variação de tempo de execução')
plt.show()
```
