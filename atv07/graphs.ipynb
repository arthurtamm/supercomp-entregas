```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
knapsack = [1, 2, 3, 4]
t_exSearch = [2.18053, 81.8085, 0, 14.3437]
t_exSearchCache = [0.00141465, 0.00157764, 0.00176917, 0.00127131]
```

```python
# Shuffle heuristic

t_shuffle_input1 = [7.667e-06, 7.686e-06, 7.872e-06, 7.752e-06, 7.467e-06]
t_shuffle_input2 = [7.458e-06, 7.267e-06, 7.338e-06, 7.332e-06, 7.356e-06]
t_shuffle_input3 = [7.726e-06, 8.223e-06, 7.545e-06, 7.528e-06, 8.11e-06]
t_shuffle_input4 = [7.257e-06, 7.858e-06, 8.053e-06, 7.313e-06, 7.331e-06]

t_shuffle = [np.mean(t_shuffle_input1), np.mean(t_shuffle_input2), np.mean(t_shuffle_input3), np.mean(t_shuffle_input4)]
```

```python
# Prob heuristic

t_prob_input1 = [8.327e-06, 8.085e-06, 7.497e-06, 8.153e-06, 8.115e-06]
t_prob_input2 = [8.367e-06, 8.326e-06, 7.787e-06, 8.711e-06, 8.502e-06]
t_prob_input3 = [7.83e-06, 7.906e-06, 8.429e-06, 8.46e-06, 7.821e-06]
t_prob_input4 = [7.251e-06, 8.104e-06, 7.328e-06, 7.977e-06, 8.071e-06]

t_prob = [np.mean(t_prob_input1), np.mean(t_prob_input2), np.mean(t_prob_input3), np.mean(t_prob_input4)]
```

```python
max_values = [601, 870, 3320, 648]
knapsack_capacities = [380, 239, 777, 600]
```

```python
# Shuffle heuristic

shuffle_max_values_input1 = [265, 262, 242, 185, 304]
shuffle_max_values_input2 = [350, 208, 288, 268, 321]
shuffle_max_values_input3 = [3320, 3320, 3320, 3320, 3320]
shuffle_max_values_input4 = [476, 566, 544, 474, 494]

shuffle_average_values = [np.mean(shuffle_max_values_input1), np.mean(shuffle_max_values_input2), np.mean(shuffle_max_values_input3), np.mean(shuffle_max_values_input4)]
shuffle_value_efficiency_ratio = [x / y for x, y in zip(shuffle_average_values, max_values)]


shuffle_final_weights_input1 = [375, 365, 370, 376, 361]
shuffle_final_weights_input2 = [238, 239, 239, 238, 238]
shuffle_final_weights_input3 = [405, 405, 405, 405, 405]
shuffle_final_weights_input4 = [597, 595, 596, 591, 594]

shuffle_average_final_weights = [np.mean(shuffle_final_weights_input1), np.mean(shuffle_final_weights_input2), np.mean(shuffle_final_weights_input3), np.mean(shuffle_final_weights_input4)]

shuffle_weight_efficiency_ratio = [x / y for x, y in zip(shuffle_average_final_weights, knapsack_capacities)]
```

```python
# Probabilistic heuristic

prob_max_values_input1 = [340, 262, 331, 305, 351]
prob_max_values_input2 = [263, 196, 381, 272, 300]
prob_max_values_input3 = [1149, 1665, 1646, 1790, 1707]
prob_max_values_input4 = [352, 416, 488, 441, 344]

prob_average_values = [np.mean(prob_max_values_input1), np.mean(prob_max_values_input2), np.mean(prob_max_values_input3), np.mean(prob_max_values_input4)]
prob_value_efficiency_ratio = [x / y for x, y in zip(prob_average_values, max_values)]

prob_final_weights_input1 = [359, 371, 380, 369, 375]
prob_final_weights_input2 = [235, 238, 239, 237, 236]
prob_final_weights_input3 = [174, 220, 192, 213, 210]
prob_final_weights_input4 = [389, 592, 594, 552, 495]

prob_average_final_weights = [np.mean(prob_final_weights_input1), np.mean(prob_final_weights_input2), np.mean(prob_final_weights_input3), np.mean(prob_final_weights_input4)]
prob_weight_efficiency_ratio = [x / y for x, y in zip(prob_average_final_weights, knapsack_capacities)]
```

```python
# Configurações gerais dos gráficos
plt.style.use('seaborn-darkgrid')
colors = ['#1f77b4', '#ff7f0e', '#2ca02c']
width = 0.25  # Largura das barras

# Gráfico 1: Tempo de Execução
def plot_execution_time():
    fig, ax = plt.subplots(figsize=(10, 6))
    x = np.arange(len(knapsack))
    
    ax.bar(x - width, t_exSearchCache, width, label='Busca Exaustiva com Cache', color=colors[0])
    ax.bar(x, t_shuffle, width, label='Heurística de Embaralhamento', color=colors[1])
    ax.bar(x + width, t_prob, width, label='Heurística Probabilística', color=colors[2])
    
    ax.set_xlabel('Entradas da Mochila')
    ax.set_ylabel('Tempo de Execução (segundos)')
    ax.set_title('Comparação de Tempo de Execução por Método')
    ax.set_xticks(x)
    ax.set_xticklabels([f'Entrada {i}' for i in knapsack])
    ax.legend()
    
    plt.tight_layout()
    plt.show()

def plot_execution_time_exhaustive():
    fig, ax = plt.subplots(figsize=(10, 6))
    x = np.arange(len(knapsack))
    
    ax.bar(x, t_exSearch, width, color=colors[0])
    
    ax.set_xlabel('Entradas da Mochila')
    ax.set_ylabel('Tempo de Execução (segundos)')
    ax.set_title('Tempo de execução da busca exaustiva sem cache')
    ax.set_xticks(x)
    ax.set_xticklabels([f'Entrada {i}' for i in knapsack])
    ax.legend()
    
    plt.tight_layout()
    plt.show()

# Chamando as funções para plotar os gráficos
plot_execution_time()
plot_execution_time_exhaustive()
```

```python
# Gráfico 2: Valor Obtido
def plot_values_obtained():
    fig, ax = plt.subplots(figsize=(10, 6))
    x = np.arange(len(knapsack))
    
    ax.bar(x - width, max_values, width, label='Valor Ótimo (Busca Exaustiva)', color=colors[0])
    ax.bar(x, shuffle_average_values, width, label='Heurística de Embaralhamento', color=colors[1])
    ax.bar(x + width, prob_average_values, width, label='Heurística Probabilística', color=colors[2])
    
    ax.set_xlabel('Entradas da Mochila')
    ax.set_ylabel('Valor Obtido')
    ax.set_title('Comparação de Valores Obtidos por Método')
    ax.set_xticks(x)
    ax.set_xticklabels([f'Entrada {i}' for i in knapsack])
    ax.legend()
    
    plt.tight_layout()
    plt.show()

plot_values_obtained()
```

```python
# Gráfico 3: Eficiência do Valor
def plot_value_efficiency():
    fig, ax = plt.subplots(figsize=(10, 6))
    x = np.arange(len(knapsack))
    
    ax.bar(x - width/2, [1]*len(knapsack), width, label='Busca Exaustiva', color=colors[0])
    ax.bar(x + width/2, shuffle_value_efficiency_ratio, width, label='Heurística de Embaralhamento', color=colors[1])
    ax.bar(x + 1.5*width, prob_value_efficiency_ratio, width, label='Heurística Probabilística', color=colors[2])
    
    ax.set_xlabel('Entradas da Mochila')
    ax.set_ylabel('Eficiência do Valor (Proporção do Ótimo)')
    ax.set_title('Comparação de Eficiência de Valor por Método')
    ax.set_xticks(x)
    ax.set_xticklabels([f'Entrada {i}' for i in knapsack])
    ax.legend()
    ax.set_ylim(0, 1.2)
    ax.set_yticklabels([f'{int(tick * 100)}%' for tick in ax.get_yticks()])
    
    plt.tight_layout()
    plt.show()

plot_value_efficiency()
```

```python
# Gráfico 4: Utilização da Capacidade da Mochila
def plot_capacity_utilization():
    fig, ax = plt.subplots(figsize=(10, 6))
    x = np.arange(len(knapsack))
    
    ax.bar(x - width, [1,1,1,1], width, label='Busca Exaustiva com Cache', color=colors[0])
    ax.bar(x, shuffle_weight_efficiency_ratio, width, label='Heurística de Embaralhamento', color=colors[1])
    ax.bar(x + width, prob_weight_efficiency_ratio, width, label='Heurística Probabilística', color=colors[2])
    
    ax.set_xlabel('Entradas da Mochila')
    ax.set_ylabel('Utilização da Capacidade (%)')
    ax.set_title('Comparação de Utilização da Capacidade da Mochila por Método')
    ax.set_xticks(x)
    ax.set_xticklabels([f'Entrada {i}' for i in knapsack])
    ax.legend()
    ax.set_ylim(0, 1.2)
    ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda y, _: '{:.0%}'.format(y)))
    
    plt.tight_layout()
    plt.show()

plot_capacity_utilization()
```

```python
# Gráfico 5: Variabilidade dos Valores Obtidos (Heurísticas)
def plot_value_variability():
    fig, axs = plt.subplots(2, 2, figsize=(12, 10))
    entries = ['Entrada 1', 'Entrada 2', 'Entrada 3', 'Entrada 4']
    shuffle_data = [
        shuffle_max_values_input1,
        shuffle_max_values_input2,
        shuffle_max_values_input3,
        shuffle_max_values_input4
    ]
    prob_data = [
        prob_max_values_input1,
        prob_max_values_input2,
        prob_max_values_input3,
        prob_max_values_input4
    ]
    
    for i, ax in enumerate(axs.flat):
        x = np.arange(1, 6)
        ax.plot(x, shuffle_data[i], marker='o', linestyle='-', color=colors[1], label='Heurística de Embaralhamento')
        ax.plot(x, prob_data[i], marker='s', linestyle='--', color=colors[2], label='Heurística Probabilística')
        ax.hlines(y=max_values[i], xmin=1, xmax=5, colors=colors[0], linestyles='dotted', label='Valor Ótimo')
        ax.set_xlabel('Execução')
        ax.set_ylabel('Valor Obtido')
        ax.set_title(entries[i])
        ax.set_xticks(x)
        ax.legend()
    
    fig.suptitle('Variabilidade dos Valores Obtidos por Heurísticas', fontsize=16)
    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()

plot_value_variability()
```
